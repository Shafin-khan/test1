Steps and Techniques for Univariate Analysis:
Descriptive Statistics:

Measures of Central Tendency:
Mean: The average value of the data.
Median: The middle value when data is ordered.
Mode: The most frequently occurring value.
Measures of Dispersion:
Range: The difference between the maximum and minimum values.
Variance: The average of the squared differences from the mean.
Standard Deviation: The square root of the variance, indicating the spread of data.
Interquartile Range (IQR): The range within which the central 50% of values fall, calculated as Q3 - Q1 (75th percentile - 25th percentile).
Frequency Distribution:

Frequency Tables: Lists the number of occurrences of each value or group of values.
Histograms: Visual representation of the frequency distribution for numerical data.
Bar Charts: Visual representation of the frequency distribution for categorical data.
Data Visualization:

Box Plots: Displays the distribution of data based on a five-number summary (minimum, first quartile, median, third quartile, maximum).
Histograms: Plots the frequency of data points in successive intervals, useful for seeing the shape of the data distribution.
Density Plots: Shows the probability distribution of a continuous variable.
Pie Charts: Used for categorical data to show the relative proportions of different categories.
Distribution Analysis:

Normality Tests: Checks if the data follows a normal distribution (e.g., Shapiro-Wilk test, Kolmogorov-Smirnov test).
Skewness and Kurtosis: Measures of the asymmetry and the peakedness of the distribution, respectively.
